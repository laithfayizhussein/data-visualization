```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
art = pd.read_csv('./artists.csv')
top_ten = art.sort_values(by='paintings', ascending=False).head(10)
top_ten 
```

# first shape Top 10 highest producing Artists

```python
plot = sns.barplot(x = 'paintings', y = 'name', data = top_ten, palette="icefire")
plot.set_title('Top 10 highest producing Artists')
plot.set(xlabel='Paintings ', ylabel='Artist Name')
print()
```

# second shape Top 10 highest producing Artists

```python
sns.relplot(x='paintings', y='name', data=top_ten, kind='line',color='red')
plot.set_title('Top 10 highest producing Artists of all time')
plot.set(xlabel='Paintings ', ylabel='Artist Name')
print()
```

# RelationShip between the painting and the year 

```python

sns.barplot(x='paintings', y='years', data=top_ten)
print()
```
